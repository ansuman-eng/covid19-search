```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import json
```

```python
sentences = []
sentenceDict = {}
```

```python
def extractText(filename):
    with open(filename) as f:
        file = json.load(f)
    bodyText = file['body_text']
    paperKey = file['paper_id']
    
    fullText = ''
    for partText in bodyText:
        fullText+=partText['text']
    fullText = fullText.split('.')
    for sentence in fullText:
        sentences.append(sentence)
        sentenceDict[sentence] = paperKey
    print("Number of sentences", len(fullText.split('.')))
    sentences.extend(fullText.split('.'))
```
